######Joins

```python
#from pyspark.sql.functions import lit,col

rawdf = spark.read.csv("/Volumes/telecom_catalog_assign/landing_zone/landing_vol/BB_Practice data/custsmodified",header=True,inferSchema=True).toDF("id","firstname","lastname","age","profession")

rawdf1=rawdf.na.drop().where("id<>'ten' and id<>'trailer_data:end of file'")
leftdf = rawdf1.select("id","firstname","lastname","age").where("id in(4000100,4000101)")
rightdf =rawdf1.select("id","profession").where("id in (4000100,4000102,4000103)")
leftdf.show(20,False)
rightdf.show(20,False)

print("innerjoin")
innerjoindf = leftdf.join(rightdf,how="inner",on="id")
innerjoindf.show(20,False)

print("leftjoin")
leftjoindf = leftdf.join(rightdf,how="left",on="id")
leftjoindf.show(20,False)

print("rightjoin")
rightjoindf = leftdf.join(rightdf,how="right",on="id")
rightjoindf.show(20,False)

print("fulljoin")
fulljoindf=leftdf.join(rightdf,how='full',on='id')
fulljoindf.show(20)

print("left semi/semi")
semijoindf=leftdf.join(rightdf,how='leftsemi',on='id')
semijoindf.show(20)

print("left anti/anti")
antijoindf=leftdf.join(rightdf,how='leftanti',on='id')
antijoindf.show(20)

print('cartesian/cross')
joindf=leftdf.join(rightdf)#without on it is cross/cartesian product (very costly and avoidable join)
joindf.show(20)#6 rows returned

print('inner by default')
joindf=leftdf.join(rightdf,on='id')#default inner join
joindf.show(20)#4000100 returned

```
